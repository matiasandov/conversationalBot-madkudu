```python
import requests
import selenium
import webbrowser
from selenium import webdriver
from selenium.webdriver.common.by import By

# to download html:
from bs4 import BeautifulSoup

# to write the files:
from datetime import date
import os

import re

import pandas as pd
import numpy as np
import time
```

```python

def extractAllText(url):
    driver = webdriver.Chrome()
    driver.get(url)
    try:
        time.sleep(5)
        #get everything inside the main div
        maindiv =  driver.find_element(By.CSS_SELECTOR, '#article-container > article')
        text = maindiv.text
        return(text)
       


    except Exception as e:
        # Handle exceptions, for example, if the element is not found
        print(f"Error: {e}")
```

```python
url = "https://support.madkudu.com/hc/en-us/articles/360040923172-How-Marketing-Uses-MadKudu"  # Replace with the URL you want to extract text from
extracted_text = extractAllText(url)
print(extracted_text)
```

```python
with open('all2.txt', 'a', encoding='utf-8') as file:
    for url in urls:
        text = extractAllText(url) + "\n"
      
        file.write(text)
```
